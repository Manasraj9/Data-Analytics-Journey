```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")
sns.scatterplot(data, x="total_bill",y="tip",hue="day",size="size",marker="o",palette="viridis")
```

```python
data = pd.read_excel("ESD.xlsx")
sns.scatterplot(data, x="Age",y="Annual Salary",hue="Department",size="Bonus %",marker="o",palette="viridis")
plt.legend(bbox_to_anchor=(0.2,0, 1.2,1))  #(x,y, height)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[23,67,45,23,12,78,45]
plt.subplot(2,2,1) #rows,columns,chartnum
plt.bar(x,y)

x = [9,10,11,12,13]
y=[67,50,66,56,82]
plt.subplot(2,2,2)
plt.scatter(x,y)

x = [9,10,11,12,13]
y=[67,50,66,56,82]
plt.subplot(2,2,3)
plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

brands = ["Oneplus","Samsung","apple","Nokia","Redmi"]
x = [22,35,30,3,10]
c = ["red","green","blue","yellow","purple"]
ex=[0,0,0,0,0.1]

plt.pie(x,labels = brands,colors=c, explode=ex, shadow = True, autopct = "%.f",startangle=180)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
# print(df)
plt.pie(grouped_by.values,labels=grouped_by.index,autopct = "%.2f" )
plt.show()
```
